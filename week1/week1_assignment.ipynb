### How `.add()` works for sets

Sets have their own method for adding items: `.add()`. Unlike lists which use `.append()`, sets use `.add()` because sets are unordered and don't allow duplicates.

```python
genres = {"Rock", "Jazz", "Pop"}

genres.add("Folk")       # Adds "Folk" — genres now has 4 items
genres.add("Rock")       # Rock already exists — nothing changes, no error
genres.remove("Pop")     # Removes "Pop" from the set

print(genres)            # {"Rock", "Jazz", "Folk"}
print(len(genres))       # 3
```

The key difference from lists:
- `my_list.append("item")` — always adds to the end of the list
- `my_set.add("item")` — adds only if the item isn't already there

---

### List slicing

Slicing lets you pull out a portion of a list using the syntax `list[start:stop]`. Python returns every item from `start` up to — but **not including** — `stop`.

```python
songs = ["A", "B", "C", "D", "E", "F", "G"]
#index:   0    1    2    3    4    5    6
```

| Expression | Result | What it means |
|------------|--------|---------------|
| `songs[0]` | `"A"` | First item (index 0) |
| `songs[2]` | `"C"` | Item at index 2 |
| `songs[-1]` | `"G"` | Last item (negative counts from the end) |
| `songs[-2]` | `"F"` | Second-to-last item |
| `songs[1:4]` | `["B", "C", "D"]` | Index 1 up to (not including) 4 |
| `songs[0:3]` | `["A", "B", "C"]` | First three items |
| `songs[4:]` | `["E", "F", "G"]` | From index 4 to the end |
| `songs[:3]` | `["A", "B", "C"]` | From the beginning up to index 3 |

```python
songs = ["Bohemian Rhapsody", "Hotel California", "Wonderwall",
         "Smells Like Teen Spirit", "Billie Jean", "Rolling in the Deep"]

print(songs[0])      # Bohemian Rhapsody  (first)
print(songs[-1])     # Rolling in the Deep  (last)
print(songs[1:4])    # ["Hotel California", "Wonderwall", "Smells Like Teen Spirit"]
print(songs[:3])     # first three songs
print(songs[3:])     # everything from index 3 onward
```

> 💡 A helpful way to remember: `songs[start:stop]` — `start` is **included**, `stop` is **excluded**.

### Numbered loops with `range()`

To print a numbered list, use `range(len(...))` to generate the index, then add 1 when printing so the count starts at 1:

```python
songs = ["Bohemian Rhapsody", "Hotel California", "Wonderwall"]

for i in range(len(songs)):
    print(f"{i + 1}. {songs[i]}")

# Output:
# 1. Bohemian Rhapsody
# 2. Hotel California
# 3. Wonderwall
```

`range(len(songs))` produces `0, 1, 2` — the valid index positions. Adding 1 in the print statement shifts the display number to start at 1.

## Quick Reference — Set Operators: `in`, `&`, and `|`

Before Task 4, here are three tools you'll use to work with sets.

### `in` — membership test

`in` checks whether a value exists inside any collection (list, set, or dictionary).

```python
genres = {"Rock", "Jazz", "Pop"}

print("Jazz" in genres)    # True  — Jazz is in the set
print("Metal" in genres)   # False — Metal is not in the set

songs = ["Bohemian Rhapsody", "Hotel California"]
print("Hotel California" in songs)   # True
```

Use `in` inside an `if` statement to make decisions:

```python
if "Jazz" in genres:
    print("This library has Jazz!")
else:
    print("No Jazz here.")
```

---

### `&` — intersection (items in **both** sets)

The `&` operator returns only the items that appear in **both** sets.

```python
my_songs      = {"Bohemian Rhapsody", "Hotel California", "Wonderwall"}
their_songs   = {"Wonderwall", "Smells Like Teen Spirit", "Hotel California"}

shared = my_songs & their_songs
print(shared)   # {"Wonderwall", "Hotel California"}
```

Think of it as: *"what do we have in common?"*

---

### `|` — union (all unique items from **either** set)

The `|` operator combines both sets and removes any duplicates.

```python
my_songs      = {"Bohemian Rhapsody", "Hotel California", "Wonderwall"}
their_songs   = {"Wonderwall", "Smells Like Teen Spirit", "Hotel California"}

combined = my_songs | their_songs
print(combined)
# {"Bohemian Rhapsody", "Hotel California", "Wonderwall", "Smells Like Teen Spirit"}
```

Think of it as: *"everything, with no repeats."*

| Operator | Name | Returns |
|----------|------|---------|
| `in` | Membership test | `True` or `False` |
| `&` | Intersection | Items in **both** sets |
| `\|` | Union | All unique items from **either** set |
